```python
from pydub import AudioSegment
import os

# Directory paths
source_directory = '/content/drive/MyDrive/Khabar'
output_directory = '/content/drive/MyDrive/Khabar/mono'

# Make sure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Specify the duration of the music at the beginning and end in milliseconds
# Adjust these values according to your needs
beginning_music_duration = 30000  # Example: first 5 seconds to remove
end_music_duration = 5000  # Example: last 5 seconds to remove

# Loop through each file in the source directory
for filename in os.listdir(source_directory):
    if not filename.endswith('.mp3'):  # Skip non-mp3 files
        continue

    # Construct the full path to the source file
    file_path = os.path.join(source_directory, filename)

    # Load the audio file
    audio = AudioSegment.from_file(file_path)

    # Trim the music from the beginning and end
    trimmed_audio = audio[beginning_music_duration:-end_music_duration]

    # Convert to mono
    mono_audio = trimmed_audio.set_channels(1)

    # Construct the output file path (change the extension to .wav)
    output_file_path = os.path.join(output_directory, os.path.splitext(filename)[0] + '.wav')

    # Export the processed audio file in WAV format
    mono_audio.export(output_file_path, format='wav')

    print(f'Processed audio file saved as: {output_file_path}')
```

```python
from pydub import AudioSegment
import os

def chunk_audio(file_path, chunk_length_ms=5000, target_sample_rate=24000):
    """Divides an audio file into chunks of a specified length and adjusts the sample rate."""
    audio = AudioSegment.from_file(file_path, format="wav")
    chunks = []

    for i in range(0, len(audio), chunk_length_ms):
        chunk = audio[i:i+chunk_length_ms]
        if len(chunk) == chunk_length_ms:
            # Adjust the sample rate of the chunk
            chunk = chunk.set_frame_rate(target_sample_rate)
            chunks.append(chunk)
        else:
            # If the chunk is less than chunk_length_ms, we discard it
            print(f"Discarded a chunk less than {chunk_length_ms} ms long")

    return chunks

def save_chunks(chunks, output_directory, global_counter):
    """Saves audio chunks to disk with global sequential naming."""
    for chunk in chunks:
        chunk_name = f"{output_directory}/{global_counter}.wav"
        chunk.export(chunk_name, format="wav")
        print(f"Exported {chunk_name}")
        global_counter += 1
    return global_counter

# Path to the directory containing the mono audio files
mono_directory = '/content/drive/MyDrive/Khabar/mono'

# New output directory for chunks
output_directory = '/content/drive/MyDrive/Khabar/mono/wavs'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Global sequential file naming
global_file_counter = 1

# Process each file in the mono directory
for filename in os.listdir(mono_directory):
    if filename.endswith('.wav'):
        file_path = os.path.join(mono_directory, filename)

        # Chunk the audio file and adjust the sample rate
        chunks = chunk_audio(file_path)

        # Save the chunks with global sequential naming and update the counter
        global_file_counter = save_chunks(chunks, output_directory, global_file_counter)

```

```python
import os
from pydub import AudioSegment
import io
from google.cloud import speech_v1 as speech
import csv

# Assuming GOOGLE_APPLICATION_CREDENTIALS is already set

# Instantiates a client
client = speech.SpeechClient()

# Directory containing the audio chunks
wavs_directory = '/content/drive/MyDrive/Khabar/mono/wavs'

# Path for the metadata.csv file
metadata_csv_path = os.path.join(wavs_directory, 'metadata2.csv')

# Function to convert stereo audio to mono
def stereo_to_mono(audio_file_path):
    sound = AudioSegment.from_wav(audio_file_path)
    sound = sound.set_channels(1)
    return sound.raw_data

# Function to transcribe a single audio file
def transcribe_file(speech_file):
    content = stereo_to_mono(speech_file)  # Convert to mono before transcription

    audio = speech.RecognitionAudio(content=content)
    config = speech.RecognitionConfig(
        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=24000,
        language_code="fa-IR"  # Setting language to Persian (Farsi)
    )

    # Detects speech in the audio file
    response = client.recognize(config=config, audio=audio)

    # Concatenate the results from the possible multiple parts
    transcription = ' '.join(result.alternatives[0].transcript for result in response.results)
    return transcription

# Create or overwrite the metadata.csv file and prepare it for data entry
with open(metadata_csv_path, mode='w', newline='', encoding='utf-8') as metadata_file:
    writer = csv.writer(metadata_file)
    writer.writerow(['audio_name', 'transcription'])  # Column headers

    # Define the range of files to process
    start_file = 4411
    end_file = 4527

    # Process each WAV file in the directory within the specified range
    for filename in os.listdir(wavs_directory):
        if filename.endswith('.wav'):
            # Extract the file number and convert it to integer
            file_number = int(filename.split('.')[0])
            # Check if the file is within the specified range
            if start_file < file_number <= end_file:
                print(f"Transcribing {filename}...")
                transcription = transcribe_file(os.path.join(wavs_directory, filename))
                writer.writerow([filename, transcription])
                print(f"Finished transcribing {filename}")

print("Selected transcriptions have been completed and saved to metadata.csv.")
```

```python
from pydub import AudioSegment
from google.cloud import speech
import io
import os
import csv

# Define the directory to save the converted mono audios
mono_audio_dir = "/content/drive/MyDrive/Khabar/mono/wavs"

# Ensure the mono audio directory exists
os.makedirs(mono_audio_dir, exist_ok=True)

# Function to convert stereo audio to mono and save it in the specified directory
def convert_stereo_to_mono(audio_path, target_dir):
    audio_stereo = AudioSegment.from_wav(audio_path)
    audio_mono = audio_stereo.set_channels(1)
    mono_filename = os.path.basename(audio_path).replace('.wav', '_mono.wav')
    mono_audio_path = os.path.join(target_dir, mono_filename)
    audio_mono.export(mono_audio_path, format="wav")
    return mono_audio_path

# Function to transcribe audio file
def transcribe_audio(speech_client, audio_path):
    with io.open(audio_path, 'rb') as audio_file:
        content = audio_file.read()
    audio = speech.RecognitionAudio(content=content)
    config = speech.RecognitionConfig(
        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=24000,
        language_code='fa-IR'
    )
    response = speech_client.recognize(config=config, audio=audio)
    transcript = " ".join([result.alternatives[0].transcript for result in response.results])
    return transcript

# Directory containing the audio files
audio_dir = "/content/drive/MyDrive/arman audio book/arman"
# CSV file to store the transcripts
csv_file_path = "/content/drive/MyDrive/arman audio book/arman/metadata.csv"

# Instantiate the Google Speech client
speech_client = speech.SpeechClient()

# Open or create the CSV file
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Clip Name", "Transcript"])  # Write the header

    # Process each audio file in the directory
    for filename in os.listdir(audio_dir):
        if filename.endswith(".wav"):
            audio_path = os.path.join(audio_dir, filename)
            print(f"Processing {filename}...")
            # Convert stereo to mono and save it in the specified directory
            mono_audio_path = convert_stereo_to_mono(audio_path, mono_audio_dir)
            # Transcribe the mono audio
            transcript = transcribe_audio(speech_client, mono_audio_path)
            # Write the clip name and transcript to the CSV
            writer.writerow([filename, transcript])
            print(f"Finished {filename}")

print("All audio files processed and transcribed.")
```

```python
import csv

# Define the path to your original CSV file and the output file
input_csv_path = '/content/drive/MyDrive/arman audio book/arman/metadata.csv'  # Update this to your CSV file path
output_csv_path = '/content/drive/MyDrive/arman audio book/arman/concatenated_metadata.csv'  # Update this to your desired output path

# Open the original CSV file to read data
with open(input_csv_path, mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    # Open the output CSV file to write concatenated data
    with open(output_csv_path, mode='w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        # Skip the header row from the input file if it exists
        next(reader, None)  # Ensures we don't include the header in our processing
        # Read each row from the input CSV, concatenate values, and write to the output CSV
        for row in reader:
            clip_name = row[0]  # Assuming "Clip Name" is the first column
            transcription = row[1]  # Assuming "Transcript" is the second column
            concatenated_value = transcription + '|' + clip_name
            writer.writerow([concatenated_value])

print("Concatenation complete. The output is saved to", output_csv_path)
```
